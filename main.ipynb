````python
# Install tokenizer and model itself
import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
````

````python
# Read the markdown file
with open("lifecycle-of-reactive-effects.md", "r", encoding="utf-8") as f:
    content = f.readlines()
````

````python
# Function to translate text
def translate_text(text):
    inputs = tokenizer(text, return_tensors="pt")
    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=tokenizer.convert_tokens_to_ids("rus_Cyrl"), max_length=500
    )
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
````

````python
translated_lines = []
inside_code_block = False

for line in content:
    if line.startswith("```"):  # Detect code blocks
        inside_code_block = not inside_code_block
        translated_lines.append(line)
        continue

    if inside_code_block or re.match(r"`.*?`", line):  # Skip inline code
        translated_lines.append(line)
    else:
        translated_lines.append(translate_text(line.strip()) + "\n")
````

````python
# Save the translated file
with open("translated_lifecycle.md", "w", encoding="utf-8") as f:
    f.writelines(translated_lines)
````
